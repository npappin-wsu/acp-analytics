```python
import pandas as pd
import os, itables
```

```python
files = [
    'ACP-Households-by-County-January-December-2022.xlsx',
    'ACP-Households-by-County-January-October-2023.xlsx'
]
```

```python
columnHint = {
    'County FIPS': str
}
```

```python
# df = pd.DataFrame()

# for file in files:
#     print(file)
df = pd.concat(pd.read_excel(os.path.join('data', file), dtype=columnHint) for file in files)
```

```python
df = df.query("State == 'WA'").reset_index(drop=True)
```

```python
# This line is to fix september of 2023... it is sad.
df['County FIPS'] = df['County FIPS'].str.pad(3, side='left', fillchar='0')
df['County Name'] = df['County Name'].str.upper()
```
